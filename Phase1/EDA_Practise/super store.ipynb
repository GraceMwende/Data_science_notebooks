- Furniture category with chairs being the top seller($328449)
- office Supplies also contributes significnatly with Sorage and binders being Key sub-categories

### b)Profit Analysis

```python
sns.barplot(data=category_sales_profit,x='total_profit',y='Sub-Category', hue='Category')
plt.title('Total Profit by Sub-Category and Category')
plt.xlabel('Total Profit')
plt.ylabel('Sub-Category')
plt.legend(title='Category');
```

- Phones in the technology category not only lead in sales but aslo contribute to the most to profits($44515)

### RFM (Recency, Frequency, Monetary) Analysis
### RFM Metrics:


```python
#Calculate latest date in the dataset
latest_date = df['Order Date'].max()
latest_date
```

```python
#preparing data for RFM analysis
rfm = df.groupby('Customer ID').agg({
    'Order Date' : lambda x: (latest_date - x.max()).days,
    'Order ID' : 'nunique',
    'Sales' : 'sum'
}).reset_index()
rfm
```

```python
#Rename columns to reflect RFM metrics
rfm.columns = ['Customer ID','Recency','Frequency','Monetary']
rfm.columns
```

```python
rfm
```

- Recency: Measures how recently a customer made a purchase. For example, Customer AA-10375 made a purchase 19 days ago.
- Frequency: Indicates the number of transactions a customer has made. For instance, Customer AA-10375 has made 9 purchases.
- Monetary: Represents the total amount of money spent by a customer. For example, Customer AA-10315 has spent ~$5,563.56.

### Insights from RFM Analysis:
- customers with low recency, high frequency, and high monetary values are likely to be the most valuable customers
- RFM analysis can be used to segment customers into different groups, such as “Loyal Customers,” “Potential Loyalists,” “At-Risk Customers,” etc.
