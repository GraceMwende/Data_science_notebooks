```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
X = 20 * np.random.rand(100) -10
y = np.array([10 * x + 50 + np.random.normal(scale=30) for x in X])
df = pd.DataFrame(np.hstack((X.reshape(-1, 1), y.reshape(-1, 1))),
                 columns=['X', 'y'])
 
```

```python
fig, ax = plt.subplots()

ax.scatter(X,y);
```

```python
#Bigger plot and add another plot

fig,ax = plt.subplots(1,2, figsize=(12,8)) #1r0w 2 columns
ax[0].scatter(X,y)
ax[1].hist(X);
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X,y,s=100)
#add line plot ontop
ax.plot(X,10*X+50,c='orange');
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X,y,s=100,c='k',alpha=0.5)
ax.plot(X,10*X+50,c='orange');
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X,y,s=100,c='k',alpha=0.5)
ax.plot(X,10*X+50,c='orange')
ax.set_title('Data and Approximating Line');
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(X,y,s=100,c='k',alpha=0.5,label='true')
ax.plot(X,10*X+50,c='orange',label='approx')
ax.set_title('Data and Approximating Line')
ax.set_ylabel('Temperaturs')
ax.set_xlabel('Longitudes')
ax.set_xticks(ticks=np.arange(-12,16,4))
ax.set_yticks(ticks=np.arange(-60,310,30))
ax.hlines(0,xmin=-12,xmax=12,label='freezing') #Horizontal line
ax.legend();
```
