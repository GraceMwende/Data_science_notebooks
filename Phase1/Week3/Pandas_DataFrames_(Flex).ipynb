### Adding to a DataFrame
#### Adding rows
Here are two rows that our engineer accidentally left out of the .csv file, expressed as a python dictionary

```python
extra_rows = {
    'age': [40,30], 'sex':[1,0], 'cp': [0,0], 'trestbps': [120,30],
    'chol': [240,200],
    'fbs':[0,0], 'restecg': [1,0], 'thalach': [120,122], 'exang': [0,1],
    'oldpeak': [0.1,1.0], 'slope': [1,1], 'ca': [0,1], 'thal': [2,3],
    'target':[0,0]
}
extra_rows
```

```python
# lets first turn this into our dataframe
# we can use the .from_dict() method
#missing = pd.DataFrame.from_dict(extra_rows,orient='index') #df will show error
#ValueError: All arrays must be of the same length
missing = pd.DataFrame(extra_rows)
missing
```

```python
#Now we just need to concatentae the two DataFrames together
heart_argument = pd.concat([heart_df,missing],
                           ignore_index=True)
```
