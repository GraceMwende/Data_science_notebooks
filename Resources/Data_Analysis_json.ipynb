### Q1.a)Build  a dictionary build a  that counts the plants by family .Use key <font color='red'> family_common_name</font>

```python
plant_dict = defaultdict(int) #provides default value for a key that does not exist
for plant in plants:
  #print(plant['family_common_name'])
  name =plant['family_common_name']
  plant_dict[name] += 1

plant_dict
```

### e)Remove families whose count are less than 10

### Q4.a)Check author whose bibibliographies start with "Sp.Pl"

```python
sps = []
for plant in plants:
  try:
    if plant['bibliography'].startswith('Sp. Pl'):
      sps.append(plant['author'])
  except:
      continue

sps[:3] #geeting first three authors
```

### b)Get count for each author

```python
# get unique authors
uniq_auth = list(set(sps))
uniq_auth
```

```python
#get count for unique author
counts =[]
for author in uniq_auth:
  counts.append(sps.count(author))
counts
```
