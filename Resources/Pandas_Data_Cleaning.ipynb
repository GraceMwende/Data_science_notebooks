What do we want to do about these missing values?

Recall that there are two general strategies for dealing with missing values:

1. Fill in missing values (either using another value from the column, e.g. the mean or mode, or using some other value like "Unknown")
2. Drop rows with missing values

```python
heroes_df['Publisher'].mode()[0] # Marvel Comics
heroes_df['Publisher'].value_counts(dropna=False,normalize=True) ##Marvel appears 52 % times followed closely by DC comics ate 28.9%
#missing values 2% of the dataset
"""
Let's take a look at those examples. We have Chuck Norris(An actual person),Katniss Everdeen (a charatcer from the Hunger Gammes book and movies
), and Flash Gordon (a charcterfrom comics published by king Features Syndicate).It doesnt seem very sensible to fill in the most common publisher(to say that these are all Marvel Comics),or
put them all in the same 'other' basket.

Also since the publisher is the only relevant variable for this question, we want to be more careful that the information encoded there
is meaningful and useful.We don't want to just haphazardly fill it in order to be able to run a statistical test or model

Therefore let's go with option 2, and drop all of the rows where the publisher is missing. With this option, we reduce the possibility of changing the meaningfulness of the data

If we were to choose option1  we should need values to replace the NaN's.We cannot use random values to fill in the Nan;s since we dont know what the distribution of the varuous publishers are. nor
do we know the frequency of each publisher. So if we replaced the Nan's with a random pusblisher,depedning on the comic book hero, we could be over-representing or under-representing a particular publisher

"""
```

Q3.What is the relationship between height and number of superpowers? And does this differ based on gender?*
Unlike the previous question, we won't be able to answer this with just `heroes_df`, since information about height is contained in `heroes_df`, while information about superpowers is contained in `powers_df`.

we need to do a join

```python
#check how heroes_df looks
heroes_df.head()
```

```python
##check how powers_df looks
powers_df.head() #both have names but our names are spread throught the dataset
```

```python
#Lets transpose and reset index so that we have names as a column
powers = powers_df.T
powers.reset_index(inplace=True)
powers
```

```python
#rename index column to name
powers.rename(columns={'index':'name'},inplace=True)
powers
```

```python
#combine the dataframes
heroes_and_powers_df = heroes_df.merge(powers, on='name',how='inner')
heroes_and_powers_df
```

```python
# Run this cell without changes

# Confirms you have created a DataFrame with the specified name
assert type(heroes_and_powers_df) == pd.DataFrame

# Confirms you have the right number of rows
assert heroes_and_powers_df.shape[0] == 647

# Confirms you have the necessary columns
# (If you modified the value of powers_df along the way, you might need to
# modify this test. We are checking that all of the powers are present as
# columns.)
assert [power in heroes_and_powers_df.columns for power in powers_df.index]
# (If you modified the value of heroes_df along the way, you might need to
# modify this as well. We are checking that all of the attribute columns from
# heroes_df are present as columns in the joined df)
assert [attribute in heroes_and_powers_df.columns for attribute in heroes_df.columns]
```

```python
#Aggregate the number of superheroes by superhores
#create a column called powwercount that has the count of each power
heroes_and_powers_df['Power Count'] =  sum(
    [heroes_and_powers_df[power_name] for power_name in powers_df.index]
    )
heroes_and_powers_df.head()
```

```python
#answer the question we plot height by powercount
fig, ax = plt.subplots(figsize=(16, 8))

ax.scatter(
    x=heroes_and_powers_df["Height"], y=heroes_and_powers_df["Power Count"], alpha=0.3
)

ax.set_xlabel("Height (cm)")
ax.set_ylabel("Number of Superpowers")
ax.set_title("Height vs. Power Count");
###np.corrcoef(heroes_and_powers_df["Height"],heroes_and_powers_df["Power Count"])[0][1]
```

```python
# we cant have negative heights
heroes_and_powers_df[heroes_and_powers_df['Height']<0].sample(5,random_state=1)
```

```python
#using query
heroes_and_powers_df.query('Height<0').sample(5, random_state=1)
```

It looks like not only are those heights negative, those weights are negative also, and all of them are set to exactly -99.0.

It seems like this data source probably filled in -99.0 as the height or weight whenever it was unknown, instead of just leaving it as NaN.

Depending on the purpose of the analysis, maybe this would be a useful piece of information, but for our current question, let's go ahead and drop the records where the height is -99.0. We'll make a new temporary dataframe to make sure we don't accidentally delete anything that will be needed in a future question.

```python
cleaned_df = heroes_and_powers_df.query('Height !=-99')
cleaned_df
```

```python
#redo without the negative heights
#answer the question we plot height by powercount
fig, ax = plt.subplots(figsize=(16, 8))

ax.scatter(
    x=cleaned_df["Height"], y=cleaned_df["Power Count"], alpha=0.3
)

ax.set_xlabel("Height (cm)")
ax.set_ylabel("Number of Superpowers")
ax.set_title("Height vs. Power Count");
np.corrcoef(cleaned_df["Height"],cleaned_df["Power Count"])[0][1]
#not much of a relationship
```

Now we can go on to answering the second half of question 2: *And does this differ based on gender?*

To indicate multiple categories within a scatter plot, we can use color to add a third dimension:

```python
fig, ax = plt.subplots(figsize=(16,8))

cleaned_df_male = cleaned_df.query('Gender=="Male"')
cleaned_df_female = cleaned_df.query('Gender=="Female"')
cleaned_df_Other =  cleaned_df.query('Gender!="Female" & Gender!="Male"')
ax.scatter(
    x=cleaned_df_male["Height"],
    y=cleaned_df_male["Power Count"],
    alpha=0.5,
    color='cyan',
    label='Male'
)
ax.scatter(
    x=cleaned_df_female['Height'],
    y=cleaned_df_female['Power Count'],
    alpha=0.5,
    color='gray',
    label='Female'
)
ax.scatter(
    x=cleaned_df_Other['Height'],
    y=cleaned_df_Other['Power Count'],
    alpha=0.5,
    color='yellow',
    label='Other'
)

# Customize appearance
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Number of Superpowers")
ax.set_title("Height vs. Power Count")
ax.legend();
```

It appears that there is still no clear relationship between count of powers and height, regardless of gender. We do however note that "Male" is the most common gender, and that male superheroes tend to be taller, on average.

```python
#
# Start by dropping rows and columns that we know are irrelevant

#only  rows with relevant publisher
publishers_df = heroes_and_powers_df.query('Publisher=="Marvel Comics" | Publisher=="DC Comics" ')
#group by publisher then count sum
publishers_df = publishers_df.groupby('Publisher').sum(numeric_only=True)
publishers_df
```

```python
#remove the height and weight column since we only need the superpowers that is from Agility of index 2 going forward
#publishers_df.loc[:,'Agility':] ##works as well
publishers_df = publishers_df.iloc[:,2:-1] #we not including the power count column we added
publishers_df
```

```python
#from how output should look like we need to transpose
publishers_df = publishers_df.T
publishers_df
```

```python
#name the index as superwor instead of publisher
publishers_df.index.name='Superpower Name'
publishers_df
```

```python
publishers_df.reset_index(inplace=True)
publishers_df
```

```python
#our question asked for top 5 most common superpowers
#top for marvel comics
marvel_common = publishers_df.drop('DC Comics',axis=1)
marvel_common = marvel_common.sort_values(by="Marvel Comics", ascending=False)[:5]
marvel_common
```

```python
#top for DC comics
DC_common = publishers_df.drop('Marvel Comics',axis=1)
DC_common = DC_common.sort_values(by="DC Comics", ascending=False)[:5]
DC_common
```

```python
#plot a bachart for the both the Marvel and the DC comics
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(14,8))

ax1.bar(marvel_common['Superpower Name'],marvel_common['Marvel Comics'])
ax2.bar(DC_common['Superpower Name'],DC_common['DC Comics'])


ax1.set(ylabel='Count of superheroes',title='Frequency of Top Superpowers in Marvel Comics')
ax1.set(ylabel='Count of superheroes',title='Frequency of Top Superpowers in DC Comics');


```
